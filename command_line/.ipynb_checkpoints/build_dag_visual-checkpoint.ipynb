```python
importlib.reload(sfs)
warnings.filterwarnings("ignore")
file_struct = sfs.SISEPUEDEFileStructure(initialize_directories = False,)

# read in DAG adjacency
fp_adj = os.path.join(file_struct.dir_ref_metadata, "sisepuede_subsector_adjacency.csv")
df_adj = pd.read_csv(fp_adj)
```

```python
subsecs = [x for x in df_adj.columns if (x != "subsector")]
adj = np.array(df_adj[subsecs])
w = np.where(adj > 0)

subsec_names = list(df_adj["subsector"])

tups = list(zip(*w))
tups = [
    (subsec_names[x[0]], subsec_names[x[1]]) 
    for x in tups if (x[0] != x[1])
] # drop self-adjacency
g = igraph.Graph.TupleList(tups, directed = True)
```
