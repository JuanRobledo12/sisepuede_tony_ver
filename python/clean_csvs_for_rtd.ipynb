## Combine variable CSVs for move to ``model_variable`` class in ``model_attributes``

```python


path = pathlib.Path(sa.dir_attribute_tables)
path_read = path.parents[2].joinpath("docs_new", "source", "csvs")

regex_all_cats = re.compile("table_varreqs_by_category_(.*).csv")
regex_partial_cats = re.compile("table_varreqs_by_partial_category_(.*).csv")


all_csvs = [
    x for x in os.listdir(path_read) 
    if (
        (regex_all_cats.match(x) is not None)
        or (regex_partial_cats.match(x) is not None)
    )
]


dict_tables = {}
field_drop = ""

dict_fields = {
    "Internal Model Variable": (0, int),
    "Default LHS Scalar Minimum at Final Time Period": (1, float),
    "Default LHS Scalar Maximum at Final Time Period": (1, float),
    "Dashboard": ("", str),
    "Reference": ("", str),
    "Data Link": ("", str)
}
def_val = 0

ordering = "append"
force_default_override_if_exists = False

ord_by_cat = [
    "Variable Type",
    "Variable",
    "Information",
    "Variable Schema",
    "Dashboard",
    "Notes",
    "Data Link",
    "Reference",
    "Internal Model Variable",
    "Default LHS Scalar Minimum at Final Time Period",
    "Default LHS Scalar Maximum at Final Time Period"
]

ord_by_partial_cat = [
    "Variable Type",
    "Variable",
    "Information",
    "Variable Schema",
    "Dashboard",
    "Categories",
    "Notes",
    "Data Link",
    "Reference",
    "Internal Model Variable",
    "Default LHS Scalar Minimum at Final Time Period",
    "Default LHS Scalar Maximum at Final Time Period"
]



dict_tables = {}
field_categories = "Categories"
val_all_cats = "all" # flag for model_variable

for csv in all_csvs:
    #print(csv)
    
    fl = pd.read_csv(os.path.join(path_read, csv))#, skipinitialspace = True)
    
    match_all = regex_all_cats.match(csv)
    match_partial = regex_partial_cats.match(csv)
    
    if match_all is not None:
        if field_categories not in fl.columns:
            fl[field_categories] = "all"
            
        grps = match_all.groups()[0]

    elif match_partial is not None:
        grps = match_partial.groups()[0]
        
    else:
        print(f"{csv} has no match")
        

    (
        dict_tables.update({grps: [fl]})
        if grps not in dict_tables.keys()
        else dict_tables[grps].append(fl)
    )
    
    
    
```

###   Loop over tables to concatenate

```python
def build_fn_out(
    grp: str,
) -> str:
    out = f"variable_definitions_{grp}.csv"
    
    return out


field_default_value = "Default Value"
fields_drop = ["Internal Model Variable", "Data Link", "Dashboard"]

ord_by_partial_cat_front = [
    "Variable Type",
    "Variable",
    "Information",
    "Variable Schema",
    "Categories",
    # "Notes",
    #"Dashboard",
    # "Data Link",
    "Reference",
    #"Internal Model Variable",
    "Default Value",
    "Default LHS Scalar Minimum at Final Time Period",
    "Default LHS Scalar Maximum at Final Time Period",
    "Simplex Group"
]

ord_by_partial_cat_end = [
    "Notes"
]


i = 0
dict_tables_new = {}
for k, v in dict_tables.items():
    
    df = pd.concat(v, axis = 0)
    if field_default_value not in df.columns:
        df[field_default_value] = 0
    
    # drop
    fields_drop_cur = [x for x in df.columns if x in fields_drop]
    if len(fields_drop_cur) > 0:
        df.drop(
            fields_drop_cur,
            axis = 1,
            inplace = True
        )
        
    # fields to order
    fields_fill = [
        x for x in df.columns 
        if x not in ord_by_partial_cat_front + ord_by_partial_cat_end
    ]
    fields_fill.sort()
    
    # order the fields 
    fields_ord = ord_by_partial_cat_front + fields_fill + ord_by_partial_cat_end
    df = (
        df[fields_ord]
        .sort_values(by = ["Variable Type", "Variable"])
        .reset_index(drop = True)
    )
    
    # export
    fp_out = path_read.joinpath(build_fn_out(k))
    df.to_csv(
        fp_out,
        index = None,
        encoding = "UTF-8",
    )
    
    dict_tables_new.update({k: df})
    print(f"Group {k} done.")
    
    
    
    
    
    
```
